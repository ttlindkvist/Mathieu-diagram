```python
#https://www.learnpyqt.com/courses/graphics-plotting/plotting-matplotlib/
class MplCanvas(FigureCanvasQTAgg):
    def __init__(self, parent=None, width=5, height=4, dpi=100):
        fig = Figure(figsize=(width, height), dpi=dpi)
        self.axes = fig.add_subplot(111)
        super(MplCanvas, self).__init__(fig)


# class MainWindow(QtWidgets.QMainWindow):
#     def __init__(self, *args, **kwargs):
#         super(MainWindow, self).__init__(*args, **kwargs)

#         # Create the maptlotlib FigureCanvas object, 
#         # which defines a single set of axes as self.axes.
#         self.sc = MplCanvas(self, width=5, height=4, dpi=100)
#         self.setCentralWidget(self.sc)

#         self.show()
#     def plot(self, xs, ys, *args, **kwargs):
#         self.sc.axes.plot(xs, ys, *args, **kwargs)
#     def clearWindow(self):
#         self.sc.axes.cla()

# if __name__ == '__main__':
#     app = QApplication.instance()
#     if app is None:
#         app = QApplication(sys.argv)
#     else:
#         print('QApplication instance already exists: %s' % str(app))
    
#     app.aboutToQuit.connect(app.deleteLater)

#     sc = MplCanvas(width=5, height=2, dpi=100)
#     layout = QVBoxLayout()
#     layout.addWidget(QPushButton('Clear'))
#     layout.addWidget(QPushButton('Redraw'))
#     sc.setLayout(layout)
#     sc.show()

#     sc.axes.plot([0,1,2,3,4], [10,1,20,3,40], c='k')
#     sc.axes.plot([5,6,7,8,9], [10,1,20,3,40])
    
#     app.exec_()



class Window(QtWidgets.QMainWindow):
    """PyQt app that closes successfully in Spyder.
    """
    def __init__(self):
        super().__init__()
        self.setGeometry(200, 100, 400, 300)
        self.button()

    def button(self):
        btn = QtWidgets.QPushButton('Quit', self)
        btn.setGeometry(150, 125, 100, 50)
        btn.clicked.connect(self.quitApp)

    def quitApp(self):
        QtCore.QCoreApplication.instance().quit()

if __name__ == '__main__':
    # app = 0
    app = QApplication(sys.argv)
    app.aboutToQuit.connect(app.deleteLater)
    win = Window()
    win.show()
    app.exec_()
```
